```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
savingsTiempos = pd.read_csv('tiempo.csv')
```

```python
#TEST 0 FUERZA BRUTA Y BACKTRACKING
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

savingsTiempos['SAV/N^3'] = savingsTiempos['tiempo'] / savingsTiempos['n']**3;
ax1 = sns.lineplot(x="n", y="SAV/N^3", data=savingsTiempos.reset_index(), label="Saving", ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='Líneas', loc='upper right')

#plt.ylim(-0.05,0.05);

plt.ylabel(r"F(G,N) = $\frac{tiempo(G,N)}{N^3}$", fontsize=16);
plt.xlabel("N", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```
